```python
# Read the Excel file into a DataFrame
df = pd.read_excel('main.xls')

# Convert 'Principal Balance' to millions
df['Principal Balance (Millions)'] = df['Principal Balance'] / 1e6

# Create a pivot table
pivot_table = df.pivot_table(
    values=['Principal Balance', 'Principal Balance (Millions)'],
    index='PRD',
    aggfunc={'Principal Balance': 'count', 'Principal Balance (Millions)': 'sum'},
    margins=True,  # Add totals
    margins_name='Total'  # Name for the totals row
)

# Rename the columns
pivot_table = pivot_table.rename(columns={
    'Principal Balance': 'Count of PRD',
    'Principal Balance (Millions)': 'Sum of Principal Balance (Millions)'
})

# Display the pivot table
print(pivot_table)
```
