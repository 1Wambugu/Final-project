```python
#Loan Amounts Trend over the years
data = {
    'Activation Year': [2010, 2011, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Sum of Loan Amount': [
        55000, 100000, 275000, 2353000, 5272000, 13041000, 3255000, 15273856, 73106139,
        181404958, 554860896, 2466031956, 7077547195
    ]
}

df = pd.DataFrame(data)

# Convert 'Sum of Loan Amount' to thousands
df['Sum of Loan Amount'] = df['Sum of Loan Amount'] / 1000

# Calculate descriptive statistics
descriptive_stats = df['Sum of Loan Amount'].describe()

# Plotting the trend over the years
plt.figure(figsize=(12, 6))

# Line plot for the sum of loan amounts over the years
sns.lineplot(x='Activation Year', y='Sum of Loan Amount', data=df, marker='o', label='Loan Amount Trend')

# Highlighting points of interest (e.g., significant growth years)
plt.scatter([2020, 2021, 2022, 2023], [181404.958, 554860.896, 2466031.956, 7077547.195], color='red', label='Significant Growth Years', s=100)

plt.xlabel('Activation Year')
plt.ylabel('Sum of Loan Amount (in thousands)')
plt.title('Loan Amount Trend Over the Years')
plt.legend()
plt.show()

# Displaying descriptive statistics
print("\nDescriptive Statistics for Loan Amounts:")
print(descriptive_stats)
```
